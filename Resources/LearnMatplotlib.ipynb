## Subplots and Legends

```python
days=[1,2,3,4,5]
original_price=[10,20,25,30,40]
predicted_price=[8,21,25,28,38]
```

```python
plt.plot(days, original_price, 'r-o', label='Original Price')
plt.plot(days, predicted_price, 'b-o', label='Predicted Price')
plt.legend()
plt.title('Plot with Legends');
plt.xlabel("Weekdays");
plt.ylabel("Price");
```

## Subplots

```python
# First Subpot - Original Price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'r-o', label='Original Price')
plt.legend()
plt.title('Original Price Over Time');
plt.xlabel("Weekdays");
plt.ylabel("Price");


#Second Subplot - Predicted Price
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'b-o', label='Predicted Price')
plt.legend()
plt.title('Predicted Price Over Time');
plt.xlabel("Weekdays");
plt.ylabel("Price");

#Third Subplot - Difference between Original Price and Predicted Price
plt.subplot(3, 1, 3)
price_difference=[o - p for o, p in zip(original_price, predicted_price)]
plt.bar(days, price_difference, color='purple', label='Price Difference')
plt.legend()
plt.title(' Difference between Original Price and Predicted Price');
plt.xlabel("Days");
plt.ylabel("Price Difference");
plt.tight_layout()
```

## Saving the image

```python
plt.plot(days, original_price, 'r-o', label='Original Price')
plt.plot(days, predicted_price, 'b-o', label='Predicted Price')
plt.legend()
plt.title('Plot with Legends');
plt.xlabel("Weekdays");
plt.ylabel("Price");
plt.savefig('example_prediction.png', dpi=300) #save the image
```
